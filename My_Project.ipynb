```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Read CSV file/ Importing dataset
df=pd.read_csv('social media influencers-youtube(actual).csv')
```

```python
#Chechking the data
#df.head()
```

```python
#Chechking the data
#df.tail()
```

```python
#Renaming column name for better knowledge
df.rename(columns={'Views (Avg.) ==0':'Views'},inplace=True)
df.rename(columns={'Likes (Avg.)':'Likes'},inplace=True)
df.rename(columns={'Comments (Avg.)':'Comments'},inplace=True)
df.rename(columns={'Followers':'Subscriber'},inplace=True)
df.rename(columns={'youtuber name':'Youtuber name'},inplace=True)
```

```python
df
```

```python
#Drop unnecessary column
df.drop(columns=['Category-2'],inplace=True)
df
```

```python
#Handlling null data
new_df=df.fillna({
    'Country':'Not-Reveal'
})
new_df
```

```python
#Conversion of data to get required information
def convert_into_int(value):
    if 'K' in value:
        return int(float(value.replace('K',''))*1000)
    elif 'M' in value:
        return int(float(value.replace('M',''))*1000000)
    else:
        return  int(value)
new_df['Comments']=new_df['Comments'].apply(convert_into_int)    
new_df['Likes']=new_df['Likes'].apply(convert_into_int)
new_df['Views']=new_df['Views'].apply(convert_into_int)    
new_df['Subscriber']=new_df['Subscriber'].apply(convert_into_int)
new_df
```

```python
new_df.describe()
```

```python
#Replacing null value as NaN
new_df=new_df.replace(0,np.NaN)
new_df
```

```python
#Handelling null data
new_df1=new_df.fillna({	
    'Likes':4.360563e+04,
    'Comments':2074.210000
})
```

```python
new_df1
```

```python
#new_df1['Comments (Avg.)']=(new_df1['Comments (Avg.)']/1000).astype(str)+'k'
#new_df1['Likes (Avg.)']=(new_df1['Likes (Avg.)']/1000).astype(str)+'k'
#new_df1
```

```python
#Handelling null data
new_df1=new_df1.fillna({
    'Category':'Creator'
})
new_df1
```

```python
new_df1['Dislikes']=new_df1['Likes']*0.16
```

```python
new_df2=new_df1
new_df2
```

```python
#Saving of file
new_df2.to_csv('social media influencers-youtube(Final).csv')
```

### Data Visualisation
<pre>
For better data representation we use seaborn library
</pre>

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

<pre>
It is impossible to plot all country,youtuber channel,catagory etc subscriber so we filter top 5 among them
</pre>

```python
top_5_country = new_df2['Country'].value_counts().head(5).index
new_df2['Country'] = new_df2['Country'].apply(lambda x: x if x in top_5_country else 'Other')
```

```python
top_5_Category = new_df2['Category'].value_counts().head(5).index
new_df2['Category'] = new_df2['Category'].apply(lambda x: x if x in top_5_Category else 'Other')
```

```python
new_df2.Subscriber=new_df2['Subscriber']/1000000
```

```python
new_df2.Views=new_df2['Views']/1000000
```

```python
new_df2.Likes=new_df2['Likes']/1000000
```

```python
new_df2.Dislikes=new_df2['Dislikes']/1000000
```

```python
new_df2.Comments=new_df2['Comments']/1000000
```

```python
top_5_Youtube_channel = new_df2['Youtube channel'].value_counts().head(5).index
new_df2['Youtube channel'] = new_df2['Youtube channel'].apply(lambda x: x if x in top_5_Youtube_channel else 'Other')
```

```python
sn.barplot(x=new_df2.Country,y=new_df2.Subscriber,color="green")
plt.ylabel("Subscriber in (M)")
```

```python
sn.pointplot(x=new_df2.Category,y=new_df2.Views,hue=new_df2.Category)
plt.ylabel("Views in (M)")
```

```python
sn.boxenplot(x=new_df2['Youtube channel'],y=new_df2.Likes,hue=new_df2["Youtube channel"])
plt.ylabel('Likes in (M)')
```

```python
sn.scatterplot(x=new_df2.Likes,y=new_df2.Dislikes)
plt.ylabel("Likes in (M)")
```
